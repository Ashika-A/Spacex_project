```python
# TASK 3: Calculate the distances between a launch site to its proximities
from math import radians, sin, cos, sqrt, atan2

# Define a function to calculate distances between two coordinates using the Haversine formula
def calculate_distance_haversine(lat1, lon1, lat2, lon2):
    # Convert latitude and longitude from degrees to radians
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)
    
    # Radius of the Earth in kilometers
    radius = 6371.0
    
    # Differences in coordinates
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    
    # Haversine formula
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    
    # Calculate the distance
    distance = radius * c
    
    return distance

# Example: Calculate the distance between a launch site and a proximity
site_latitude = 28.5721  # Replace with the actual launch site latitude
site_longitude = -80.6480  # Replace with the actual launch site longitude
proximity_latitude = 28.3922  # Replace with the proximity latitude
proximity_longitude = -80.6077  # Replace with the proximity longitude

distance_km = calculate_distance_haversine(site_latitude, site_longitude, proximity_latitude, proximity_longitude)
print(f"Distance between launch site and proximity: {distance_km:.2f} km")


```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
